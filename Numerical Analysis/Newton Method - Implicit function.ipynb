## Solving non-linear equations

#### Newton's Method

Compute *implicit function* $G(x,y)$ using Newton's method.

Let $$G(x,y) \ = \ 3x^7 + 2y^5 - x^3 + y^3 - 3 = 0$$

Let $x_0 = 0$, we will implement steps of $.01$ from $x = 0$ to $x = 10$. At each step, we assume $x_0$ is a fixed parameter and consider $G$ and a function of $y$. Recall the Newton's method:
$$ y_{n+1} \ = \ y_{n} - \frac{G(x_o, y_{n})}{\frac{\partial}{\partial y} G(x_0, y_{n})}
    \ = \ y_{n} - \frac{3x_0^7 + 2y_n^5 - x_0^3 + y_n^3 - 3}{10y_n^4 + 3y_n^2}
$$


```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
def G(x,y):
    G_xy = 3*x**7 + 2*y**5 - x**3 + y**3 - 3
    return(G_xy)

def G_der(x,y):
    G_der_xy = 10*y**4 + 3*y**2
    return(G_der_xy)
```

Main function

```python
# Initialize staring parameters
x, y, h, M, N = 0, 1, .1, 100, 4

table = pd.DataFrame(columns = ['i', 'x', 'y', 'G'])
for i in range(0,M):
    x = x+h
    for j in range(0,N):
        y = y - G(x,y) / G_der(x,y)
    table.loc[i] = (i,x,y,G(x,y))
```
