## Gradient Descent Algorithm

Recall that given the objective function of the form
$$ \min_{x \in \mathbb{R}^d} \ f(x)
$$

Starting at an initial point of $x^{(0)}$, for each subsequent iteration, we update as:
$$ x^{(t+1)} \ = \ x^{(t)} - \alpha_t \nabla f(x^{(t)})
$$
and continue on until one of the following criteria is satisfied (the list is not exhaustive and is meant for illustration only):

1.  reaching the maximum number of iterations wanted
2.  $f < \varepsilon$: the loss function is small enough
3.  $\|\nabla f(x)\| < \delta$: the norm of the slope of the loss function is small enough

***

In this project, we will

1.  Write a generic function to run *gradient descent*
2.  Apply the function on these 4 objective functions:
    (a) $f = e^{x+y}$ </br>
    (b) $f = \ln(1 + e^{x+y})$ </br>
    (c) $f = x^2 + y^2$ </br>
    (d) $f = x^2 + xy + y^2$

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

`grad_desc_alg` a generic function to execute the gradient descent algorithm

```python
def grad_desc_alg(f, f_grad, x, M, alpha):
    
    loss_vector = np.zeros(M)     # Initialize loss vector
    
    for m in range(0,M):          # Train model

        loss_vector[m] = f(x)   # Current loss

        grad_eval = f_grad(x)   # Evalute gradient at current point
        x = x - alpha * grad_eval # Update x
        
    return(loss_vector)
```

Helper functions on the $4$ functions:

(a) $f = e^{x+y}$ </br>
(b) $f = \ln(1 + e^{x+y})$ </br>
(c) $f = x^2 + y^2$ </br>
(d) $f = x^2 + xy + y^2$

```python
# (a) f = exp(x+y)
def f_a(x):
    return(np.e**(x[0]+x[1]))
def f_a_grad(x):
    return(np.array([f_a(x), f_a(x)]))

# (b) f = ln(1 + exp(x+y))
def f_b(x):
    return(np.log(1 + np.e**(x[0]+x[1])))
def f_b_grad(x):
    return(np.array([f_a(x)/(1+f_a(x)), 
                     f_a(x)/(1+f_a(x))]))

# (c) f = x^2 + y^2
def f_c(x):
    return(x[0]**2 + x[1]**2)
def f_c_grad(x):
    return(np.array([2*x[0], 2*x[1]]))

# (d) f = x^2 + xy + y^2
def f_d(x):
    return(x[0]**2 + x[1]**2 + x[0]*x[1])
def f_d_grad(x):
    return(np.array([2*x[0] + x[1], 2*x[1] + x[0]]))
```

Parameters:
$$ x^{(0)} = \begin{bmatrix} 0 \\ 0 \end{bmatrix};
    \ \ \ \ \ \ 
    M = 1,000;
    \ \ \ \ \ \ 
    \alpha = .01
$$

```python
alpha = .01
M = 1000

x = np.array([0,0])
loss_a_1 = grad_desc_alg(f_a, f_a_grad, x, M, alpha)
loss_b_1 = grad_desc_alg(f_b, f_b_grad, x, M, alpha)
loss_c_1 = grad_desc_alg(f_c, f_c_grad, x, M, alpha)
loss_d_1 = grad_desc_alg(f_d, f_d_grad, x, M, alpha)
```

```python
figure(num=None, figsize=(12, 6), dpi=80, facecolor='w', edgecolor='k')

idx = np.arange(0, M)

plt.scatter(idx, loss_a_1, s = 10, color = 'blue', alpha=0.5)
plt.plot(idx, loss_a_1, '-', color = 'blue', alpha=0.5, 
         label = '$f = e^{x+y}$')
plt.scatter(idx, loss_b_1, s = 10, color = 'black', alpha=0.5)
plt.plot(idx, loss_b_1, '-', color = 'black', alpha=0.5, 
         label = '$f = \ln(1 + e^{x+y})$')
plt.scatter(idx, loss_c_1, s = 10, color = 'red', alpha=0.5)
plt.plot(idx, loss_c_1, '-', color = 'red', alpha=0.5, 
         label = '$f = x^2 + y^2$')
plt.scatter(idx, loss_d_1, s = 10, color = 'green', alpha=0.5)
plt.plot(idx, loss_d_1, '-', color = 'green', alpha=0.5, 
         label = '$f = x^2 + y^2 + xy$')

plt.title('Starting at x = (0,0)', fontsize = 20)
plt.xlabel('Iteration', fontsize = 15)
plt.ylabel('Loss function', fontsize = 15)
plt.legend()

plt.savefig('173B - HW1 - P3 - x=(0,0).png', bbox_inches='tight')
```
